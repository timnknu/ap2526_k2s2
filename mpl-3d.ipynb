```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#%matplotlib widget
%matplotlib ipympl

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

ax.scatter([1, 2, 3], [4, 4, 7], [1, 2, 2])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z axis')


plt.show()
```

```python
#%matplotlib widget
%matplotlib ipympl

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# Make data
n = 100
xs = np.linspace(0, 1, n)
ys = np.sin(xs * 6 * np.pi)
zs = np.cos(xs * 6 * np.pi)

ax.scatter(xs, ys, zs)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z axis')


plt.show()
```

```python
#%matplotlib widget
%matplotlib ipympl

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# Make data
n = 100
xs = np.linspace(0, 1, n)
ys = np.sin(xs * 6 * np.pi)
zs = np.cos(xs * 6 * np.pi)

ax.plot(xs, ys, zs)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z axis')


plt.show()
```

```python
#%matplotlib widget
%matplotlib ipympl

fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, dpi=120, figsize=(7, 7))

# Make data
n = 1000
xs = np.random.uniform(low=-1, high=1, size=n)
ys = np.random.uniform(low=-1, high=1, size=n)
zs = xs**2 + np.sin(ys * 8)

ax.plot_trisurf(xs, ys, zs)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z axis')


plt.show()
```

```python
plt.figure()
plt.plot(xs, ys, 's', color='forestgreen')
plt.show()
```
